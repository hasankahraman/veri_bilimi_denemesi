```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
data = pd.read_csv("final_data.csv")
```

```python
data.head()
```

```python
data.drop(["Unnamed: 0.1","Unnamed: 0","Id"], axis = 1, inplace = True)
data.head()
```

Label Encoding

```python
le = LabelEncoder()
```

```python
data["Species"] = le.fit_transform(data["Species"])
```

```python
data.head()
```

```python
data.to_csv("modelv2.csv")
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1], test_size=0.2)
```

```python
x_test
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

## Modelin Oluşturulması

```python
import xgboost as xgb
```

```python
xgb_cls = xgb.XGBClassifier(objective="multiclass:softmax", num_class=3, use_label_encoder=False)
```

```python
xgb_cls.fit(x_train, y_train, use_label_encoder=False)
```

```python
preds = xgb_cls.predict(x_test)
```

```python
preds
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```
